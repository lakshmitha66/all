```python
temperatures = pd.DataFrame({
    'city': ['New York', 'London', 'Paris', 'Berlin', 'Amsterdam'],
    'day1': [23, 25, 27, 26, 24],
    'day2': [22, 21, 25, 26, 23],
    'day3': [26, 25, 24, 27, 23],
    'day4': [23, 21, 22, 26, 27],
    'day5': [27, 26, 27, 24, 28]
})
temperatures
```

```python
temperatures.melt()
```

```python
temperatures.melt(id_vars=['city'])
```

```python
temperatures.melt(id_vars=['city'],
                  var_name='date',
                  value_name='temperature').sample(5)
```

```python
stocks = pd.read_csv('prices-split-adjusted.csv',
                     usecols=['date', 'symbol', 'open', 'close'],
                     parse_dates=['date'],
                     index_col='date')
stocks.head()
```

```python
stocks_small = stocks.loc['2016-02-01':'2016-03-01'].reset_index()
stocks_small.head()
```

```python
reshaped=stocks_small.head()
melted =reshaped.melt(id_vars=['symbol'])
melted
```

```python
print("Number of rows in melted table: " + str(melted.shape[0]))
```
