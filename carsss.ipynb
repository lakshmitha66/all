```python
import pandas as pd
df1 = pd.read_csv('D:/aiml/excel/Cars93.csv')
df1.info()
```

```python
df1.plot(x = 'Price', y = 'Horsepower', marker = 'o', kind = 'scatter');
```

```python
df1.head()['Width'].plot(kind = 'bar')
```

```python
df1['Width'].plot(kind = 'hist')
```

```python
import matplotlib.pyplot as plt
fig = plt.figure(figsize=[6,4])
ax = fig.add_axes([0, 0, 1, 1])
ax.hist(df1['Price'])
ax.set_title('Distribution of Price',fontsize = 17)
ax.set_ylabel('Weight',fontsize=12)
ax.set_xlabel('Horsepower',fontsize=12)
fig.show()
```
