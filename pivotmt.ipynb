```python
import pandas as pd
df1 = pd.read_csv('D:/aiml/excel/mtcars.csv')
df1.info()
```

```python
df1.plot(x = 'wt', y = 'mpg', marker = 'o', kind = 'scatter')
```

```python
df1.head()['am'].plot(kind = 'bar')
```

```python
df1['mpg'].plot(kind = 'hist')
```

```python
import matplotlib.pyplot as plt
fig = plt.figure(figsize=[6,4])
ax = fig.add_axes([0, 0, 1, 1])
ax.hist(df1['mpg'])
ax.set_title('Distribution of mpg',fontsize = 17)
ax.set_ylabel('Frequency',fontsize=12)
ax.set_xlabel('Horsepower',fontsize=12)
fig.show()
```

```python
import matplotlib.pyplot as plt
#creating an empty canvas/figure
fig = plt.figure(figsize=[4,4])
#setting axes
ax = fig.add_axes([0, 0, 1, 1])
ax.set_title('Distribution of mileage')
#plotting boxplot
ax.boxplot(df1.tail()['mpg'])
fig.show()
```
