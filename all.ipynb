```python
import pandas as pd
marks = {'M1': [50,35,98,70,100], 
            'M2': [60,60,99,50,18],  
            'M3': [80,40,50,23,15],  
            'M4': [40,20,18,35,90]}
df = pd.DataFrame(marks, index = ['Gruthi', 'Nishmitha', 'Shreya', 'Lakshmitha','Shivani'])
df
df['Total'] = df['M1'] + df['M2'] +df['M3'] + df['M4']
df
df['result'] = (df['M1'] < 35) | (df['M2'] <35) | (df['M3']<35) |(df['M4']<35)
pf_dict = {True: 'Fail', False: 'Pass'}
df['result']=df['result'].replace(pf_dict)
df
df.sort_index()
df.mask(df == 35,'just pass')
```
